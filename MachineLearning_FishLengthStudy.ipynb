```python
# 도미 생선의 길이
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]

# 도미 생선의 무게
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

# 특성 : Data의 특징
# 첫번째 도미의 특성(길이 : 25.4cm, 무게 : 242.0g)
# 두번째 도미의 특성(길이 : 26.3cm, 무게 : 290.0g)
```

```python
# 두 특성을 숫자로 보는 것 보다 그래프로 표현하면 데이터를 잘 이해할 수 있음
# 길이 : x축, 무게 : y축 -> 산점도(두 좌표의 관계)

import matplotlib.pyplot as plt
# python에서 과학 계산용 그래프를 그리는 대표적인 package : matplotlib
plt.scatter(bream_length, bream_weight)    # 차트의 값을 넣어줌
plt.xlabel('Length')
plt.ylabel('Weight')
plt.show()
# 도미 35마리를 2차원 그래프에 점으로 나타냄
```

```python
# 생선의 길이가 길수록 무게가 많이 나간다 -> 선형적
```

```python
# 빙어 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
plt.scatter(smelt_length, smelt_weight)    # 차트의 값을 넣어줌
plt.xlabel('Length')
plt.ylabel('Weight')
plt.show()
```

```python
# 이 2개의 리스트를 합쳐 AI 학습용 자료로 만들어야 함
length = bream_length + smelt_length    # 도미 길이 List + 빙어 길이 List
weight = bream_weight + smelt_weight    # 도미 무게 List + 빙어 무게 List

# 사이킷런을 이용한 AI트레이닝
# 사이킷런의 입력은 2차원 리스트로 만들어야 함
```

```python
# zip() : 나열된 리스트 각각에서 하나씩 원소를 꺼내어 반환
fish_data = [[l, w] for l, w in zip(length, weight)]    # length와 weight를 2차원 리스트로 만듬
print(fish_data)    # 출력
```

```python
# 생선 49개에 대한 정답 리스트를 만들어야함
# Machine Learning에서는 정답 데이터를 2진 분류로 Test(1 : 도미, 0 : 빙어)

fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
# 전처리 데이터가 완성되었으니, 사이킷런 패키지를 활용하여 트레이닝
from sklearn.neighbors import KNeighborsClassifier    # 사이킷런 패키지 안에, 이웃패키지 / KneighborsClassifier는 k-최근접 이웃 알고리즘
kn = KNeighborsClassifier()    # import된 객체를 생성하여 변수에 연결
```

```python
# fish_data, fish_target을 전달하여 도미를 찾기 위한 기준을 학습(Training)
kn.fit(fish_data, fish_target)
```

```python
# kn 얼마나 잘 훈련되었는지 평가 / score()
kn.score(fish_data, fish_target)    # 1.0 = 100%(정확도)
```

```python
# k-최근접 이웃 알고리즘 파악
# 어떤 데이터에 대한 답을 구할 때 주위의 다른 데이터를 보고 다수를 차지하는 것을 정답으로 출력
```

```python
kn.predict([[30,  600]])    # .predict() : 새로운 데이터의 정답을 예측
# 결론 1 -> 도미
```

```python
# k-최근접 이웃 알고리즘을 위해 준비해야할 일은 데이터를 모두 가지고 있는것이 전부임.
# 새로운 데이터에 대해 예측할 시 가장 가까운 직선거리에 어떤 데이터가 있는지 살피기만 하면 됨
# 단점 : 데이터가 아주 많은 경우 사용하기 어려움(데이터가 크기 때문에 메모리가 많이 필요함, 직선거리 계산 시간도 오래걸림)
# 객체를 메모리에 만들고 활용한다

print(kn._fit_X)
```

```python
print(kn._y)
```

```python
# 결론 k-최근접 이웃 알고리즘은 훈련결과 파일이 존재하지 않음
# 새로운 데이터가 들어오면 가장 가까운 데이터를 메모리에서 참고하여 도미, 빙어를 구분함
kn.predict([[5.0 ,15.0]])
```

```python
# 객체 생성시 n_neighbors=? 참고할 이웃 데이터 수 / 기본값은 5
kn49 = KNeighborsClassifier(n_neighbors = 49)    # 참고 데이터를 49개로 한 모델
# 가장 가까운 데이터 49개를 사용함

kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)    # 훈련결과 71.4%로 좋지 않은 성능을 보임
```
